```python
# TODO 03 - your friend ask for `California` data, filter it and export csv for him
# --> California` data
df_california = df[ df['State'] == 'California'].reset_index()
df_california = df_california.drop('index', axis = 1)
# --> Export CSV file
df_california.to_csv("Data_California")
df_california
```

```python
# TODO 04 - your friend ask for all order data in `California` and `Texas` in 2017 (look at Order Date), send him csv file
# convert 'Order Date' into datetime format
# --> Data in `California` and `Texas` in 2017
df_cali_Taxa_2017 = df[df['State'].isin(['California', 'Texas']) & (df['Order Date'].dt.year == 2017)].reset_index()
df_cali_Taxa_2017 = df_cali_Taxa_2017.drop('index', axis = 1)
# --> Export CSV file
df_cali_Taxa_2017.to_csv("Data_California_Texas_2017")
df_cali_Taxa_2017
```

```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
df_agg = df[df['Order Date'].dt.year == 2017]['Sales'].agg(['sum', 'mean', 'std']) 
print(f'In 2017, our company made total sales of {round(df_agg[0],2)}$ \
        \nThe average sales were {round(df_agg[1],2)}$, and \
        \nThe standard deviation of sales was {round(df_agg[2],2)}$')
df_agg.round(2).reset_index()
```

```python
# TODO 06 - which Segment has the highest profit in 2018
df['Year'] = df['Order Date'].dt.year
df_seg_2018 = df[df['Year'] == 2018].groupby(['Year', 'Segment'])['Profit'].sum()
segment = df_seg_2018.max()
print(f'In 2018, the consumer segment had the highest profit, totaling {round(segment,2)}$')
df_seg_2018.round(2).reset_index()
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
df6_least_sales = df[(df['Order Date'] > '2019-04-15') & (df['Order Date'] < '2019-12-31')].groupby('State')['Sales'].sum()
df6_least_sales = df6_least_sales.sort_values(ascending=True).head(5)
least_sales_states = df6_least_sales.index

print(f'Top 5 States have the least total sales are {least_sales_states[0]}, {least_sales_states[1]}, {least_sales_states[2]}, {least_sales_states[3]}, and {least_sales_states[4]}.')
df6_least_sales.reset_index()
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
df_2019 = df[df['Order Date'].dt.year == 2019].groupby(['Region'])['Sales'].sum()
df_total_sales = (((df_2019['West'] + df_2019['Central'])/ df_2019.sum())*100).round(2)
print(f'The proportion of total sales (%) in West and Central in 2019 was {df_total_sales}%.')
df_2019.round(2).reset_index()
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
Top_10_pop_pro = df[(df['Order Date'].dt.year >= 2019) & (df['Order Date'].dt.year <= 2020)].groupby('Product Name')['Product Name']\
    .count().sort_values(ascending = False).head(10)
print('Top 10 popular products in terms of number of orders')
print(Top_10_pop_pro)

print('--------------------------------------------------------------')

Total_sales = (df[(df['Order Date'].dt.year >= 2019) & (df['Order Date'].dt.year <= 2020)].groupby('Product Name')['Sales']\
    .sum().sort_values(ascending = False).head(10)).round(2)
print('Total sales during 2019 - 2020')
print(Total_sales)
```
